#### 2.2.2. Translate color from German to English

```python
def translate_color(color):
    """
    Translates the color from German to English.
    Could be optimized by translating from any color into English.
    """
    translator = Translator()
    
    color_eng = translator.translate(color, src='de').text
    
    return color_eng

colors = data['BodyColorText_new'].unique()
dic_colors = {} # translated colors
# use when the API call reports too many calls
# dic_colors = {'anthrazit': 'anthracite',
#  'beige': 'beige',
#  'blau': 'blue',
#  'bordeaux': 'bordeaux',
#  'braun': 'brown',
#  'gelb': 'yellow',
#  'gold': 'gold',
#  'grau': 'Gray',
#  'grün': 'green',
#  'orange': 'orange',
#  'rot': 'red',
#  'schwarz': 'black',
#  'silber': 'silver',
#  'violett': 'violet',
#  'weiss': 'White'}
for c in colors:
    dic_colors[c] = translate_color(c)
```

#### 2.2.4. Find closest color by comparing euclidean distance of RGB color space

Closeness is defined here by the closest euclidean distance in the RGB space.

```python
# get the vector of colors for all colors in the dataset, be careful for the "Other" color
num_target_colors = len(data_tar['color'].unique().tolist())
c_vec_target = np.asarray([np.zeros(3) for i in range(num_target_colors)]) # color vector for target
for ii, color in enumerate(data_tar['color'].unique().tolist()):
    if color == 'Other':
        c_vec_target[ii] = np.asarray([1e10, 1e10, 1e10]) # some ridiculous large value
    else:
        rgb = webcolors.name_to_rgb(color, spec='css3')
        c_vec_target[ii] = np.asarray([rgb[0], rgb[1], rgb[2]])

# get euclidian distances for each of the colors in the input dataset
input_colors = data['BodyColorText_trans'].unique().tolist()
distances = []

for color in input_colors:
       
    d = np.asarray([0 for i in range(num_target_colors)]) # one value for each target color
    for ii, target_color in enumerate(c_vec_target):
        if color == 'Anthracite':
            # not defined in css3, colorcombos.com
            y = np.asarray([56, 62, 66])
        elif color == 'Bordeaux':
            # not defined in css3, colorcombos.com
            y = np.asarray([95, 2, 31])
        else:
            rgb = webcolors.name_to_rgb(color, spec='css3')
            y = np.asarray([rgb[0], rgb[1], rgb[2]])

        x = target_color
        d[ii] = np.linalg.norm(x - y) # euclidean distance
    distances.append(d)
```

```python
# visualize it
for color, d in zip(input_colors, distances):
    # add also the closest
    closest_idx = np.argmin(d)
    plt.title(f"{color}, closest: {data_tar['color'].unique().tolist()[closest_idx]}")
    plt.bar(np.arange(num_target_colors), d)
    plt.xticks(np.arange(num_target_colors), data_tar['color'].unique().tolist(), rotation=45)
    plt.ylabel('Euclidean distance')
    plt.ylim(0, 500)
    plt.show()
```

### 2.3. Normalization of `MakeText` to **make**


The idea is to compare the similarity between the words in the target dataset and the input dataset to match the car makers. For that I will use the Jaro-Winkler distance. It is best suited for short strings such as names with the goal of comparing these two names. It compares mathematically the number of matching characters and the sequence order. A value close to 0 indicates not similar at all, and 1 indicates an exact match.

```python
makers_input = pd.Series(data['MakeText'].unique()) # makers in the input file
# makers in target file
makers_target = pd.Series(data_tar['make'].unique().tolist()) 
makers_target = makers_target.dropna() # drop the nan
makers_target_lowercase = makers_target.str.lower()

# convert to dataframe for easier comparison
df_makers = pd.DataFrame(makers_input, columns=['makers_input'])


    

def compare_makers(row, makers_target, makers_target_lowercase, threshold=0.9):
    # make input lowercase
    s = row.lower()
    # calculate jaro-winkler score between the lowercase strings 
    jw_scores = makers_target_lowercase.apply(lambda x: jellyfish.jaro_winkler(s, x))
    idx_max = np.argmax(jw_scores) # maximal score index 
    score_max = jw_scores.iloc[idx_max]
    best_match = makers_target.iloc[idx_max]
    if score_max < threshold:
        print(f"For input {row}, the best match was {best_match} with a JW score of {score_max}.")
        
    return pd.Series([row, best_match, score_max], index=['maker_input', 'best_match', 'JW score'])

threshold = 0.9
df_maker_match = df_makers['makers_input'].apply(lambda row: compare_makers(row, makers_target, makers_target_lowercase, threshold))
df_maker_match
```

```python
df_maker_match[df_maker_match['JW score'] < 1]
```

It seems as we should set the JW score limit to 0.879, then all the examples in this set will be captured correctly. Best idea is to warn the user in the end, if a vendor is classified as being an "Other" and allowing to manually adjust. This should be done so as to minimize manual efforts while retaining a high precision of this probability matching approach.

Another idea here would be to query some large car manufacturer database (or a search engine using data mining) to find new car manufacturer's names. For example find the correct spelling for the HARLEY-DAVIDSON by querying wikipedia and pulling the right spelling. This does depend on the requirements from the client. If they only want what their current dataset contains or more.

### 2.5. Get `country` from `City`

This should be a fast thing to do. There are many options, using Selenium and then reading Google Maps (or using an API directly) is an option, using bs4 and DuckDuckGo is another or for example using Geopandas, which will be shown here.

```python
# use geopandas
df_city = pd.DataFrame(data['City'].unique(), columns=['City'])

def get_country(city):
    r = geopandas.tools.geocode(city, provider='nominatim', user_agent='autogis_xx', timeout=4) # get address
    assert r.shape[0] == 1, "More than one address for that city!" # make sure only one city
    long, lat = r['geometry'].x.values[0], r['geometry'].y.values[0] # get long and lat
    locator = Nominatim(user_agent="myGeocoder") 
    coordinates = f"{lat}, {long}"
    location = locator.reverse(coordinates) # get location information
    country = location.raw['address']['country_code'] # return country code
    return country.upper()

df_city['Country'] = df_city['City'].apply(lambda x: get_country(x))
```
